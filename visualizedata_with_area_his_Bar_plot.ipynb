```python
import numpy as np
import pandas as pd
```

```python
df_canada = pd.read_excel('https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/DV0101EN/labs/Data_Files/Canada.xlsx',
                         sheet_name='Canada by Citizenship',
                         skiprows = range(20),
                         skipfooter = 2)
```

```python
df_canada.head(2)
```

```python
df_canada.drop(['AREA', 'REG', 'DEV', 'Type', 'Coverage'], axis = 1, inplace = True)
```

```python
df_canada.head()
```

```python
df_canada.rename(columns = {'OdName':'Country', 'AreaName': 'Continent','RegName':'Region'}, inplace = True)
```

```python
df_canada.head()
```

```python
df_canada.columns = list(map(str, df_canada.columns))
```

```python
#check columns labebls type
all(isinstance(column, str) for column in df_canada.columns)
```

```python
df_canada.set_index('Country', inplace = True)
```

```python
df_canada.head()
```

```python
df_canada['Total'] = df_canada.sum(axis=1)
```

```python
df_canada.columns.values
```

```python
df_canada.head()
```

```python
data_dimension = df_canada.shape
print(data_dimension)
```

```python
#lets create a list of years
years = list(map(str, range(1980, 2013)))
print(years
     )
```

```python
%matplotlib inline

import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
df_canada.sort_values(['Total'], ascending = False, axis = 0, inplace= True)

df_top5 = df_canada.head()
#transpose
df_top5 = df_top5[years].transpose()
df_top5.head()
```

```python
df_top5.index = df_top5.index.map(int)

df_top5.plot(kind='area',
             stacked = False , figsize = (20, 10))

plt.title('Immigration trend of top countries')
plt.ylabel('Number of immigrant')
plt.xlabel('Years')
plt.show()
```

```python
# option 1: scripting layer
df_top5.plot(kind='area',
           alpha=0.25,# 0-1, default value a = 0.5,
           stacked=False,
           figsize=(20,10))

plt.title('Immigration trend of top countries')
plt.ylabel('Number of immigrant')
plt.xlabel('Years')
plt.show()
```

```python
#option 2: Artist layer

ax = df_top5.plot(kind='area', alpha=0.35, stacked=False, figsize=(20,10))

ax.set_title('Immigration trend of top countries')
ax.set_ylabel('Number of immigrant')
ax.set_xlabel('Years')
```

```python
df_least5 = df_canada.tail(5)
```

```python
df_least5.columns.values
```

```python
df_least5 = df_least5[years].transpose()
```

```python
df_least5.head()
```

```python
df_least5.index = df_least5.index.map(int)

df_least5.plot(kind='area', alpha=0.25, stacked = False, figsize=(20,10))

plt.title('Less contribution')
plt.ylabel('Number of immigrants')
plt.xlabel('Years')
plt.show()
```

```python
#using artist layer

al = df_least5.plot(kind='area', alpha = .55, stacked = False, figsize=(20,10))
al.set_title('Least Contribution')
al.set_ylabel('Number of immigrants')
al.set_xlabel('years')
```

```python
####  histogram
#what is the frequency distribution of the numbers of new immogrants from the various countries
# to canda in 2013
df_canada['2013'].head()
```

```python
#np.histogram return 2 values

count, bin_edges = np.histogram(df_canada['2013'])

print(count) #frequency count
print(bin_edges) #btn ranges, default  = 10 bins
```

```python
df_canada['2013'].plot(kind='hist', figsize=(8,5))

plt.title('Histogram of immigration from 195 countries in 2013')
plt.ylabel('Number of countries')
plt.xlabel('Number of immigrants')
```

```python
count, bin_edges = np.histogram(df_canada['2013'])
df_canada['2013'].plot(kind='hist', figsize=(8,5), xticks = bin_edges)
plt.title('Histogram of immigration from 195 countries in 2013')
plt.ylabel('Number of countries')
plt.xlabel('Number of immigrants')
plt.show()
```
