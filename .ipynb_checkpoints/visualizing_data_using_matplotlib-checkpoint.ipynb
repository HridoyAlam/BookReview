```python
df_canada = pd.read_excel('https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/DV0101EN/labs/Data_Files/Canada.xlsx',
                         sheet_name='Canada by Citizenship',
                         skiprows = range(20),
                         skipfooter = 2)
```

```python
df_canada.head(2)
```

```python
df_canada.columns.values
df_canada.columns.tolist()
df_canada.index.tolist()

print(type(df_canada.columns.tolist()))
```

```python
df_canada.drop(['AREA','REG','DEV','Type','Coverage'],axis=1,inplace = True)
```

```python
df_canada.rename(columns ={'OdName':'Country', 'AreaName':'Continent', 'RegName':'Region'}, inplace = True)
```

```python
df_canada.columns.values
```

```python
df_canada['Total'] = df_canada.sum(axis=1)
df_canada.columns.values
```

```python
df_canada.set_index('Country', inplace=True)
```

```python
df_canada.columns = list(map(str,df_canada.columns))
```

```python
years = list(map(str, range(1980, 2014)))
```

```python
haiti = df_canada.loc['Haiti',years]
```

```python
haiti.plot()
```

```python
haiti.index = haiti.index.map(int) # lets change the index values of haiti into the type of integers
haiti.plot(kind='line')

plt.title('Immigration from haiti')
plt.ylabel('Number of immigrants')
plt.xlabel('Years')
plt.show()
```

```python
china = df_canada.loc['China',years]
china.head()
```

```python
india = df_canada.loc['India',years]
india.head()
```

```python
#get india and china data set together
df_CI = df_canada.loc[['India', 'China'],years]
df_CI.head()
```

```python
df_CI.plot(kind='line')
```

```python
#it doesn't look good, so need to change the row and columns using pandas transpose() method
df_CI = df_CI.transpose()
df_CI.head()
```

```python
df_CI.plot(kind='line')
```

```python
df_CI.index = df_CI.index.map(int)
df_CI.plot(kind = 'line')
plt.title('Immigration from China and india')
plt.xlabel('Years')
plt.ylabel('Numbers')
plt.show()
```
